### 映射
包含三种操作：  
- replace()函数：替换元素
- 最重要：map()函数:新建一列
- rename() 函数:替换索引

```python
# replace替换元素
```

```python
df = pd.DataFrame(np.random.randint(0, 5, (5, 3)))
df
```

```python
# 替换单个值
df.replace(0, -99)
```

```python
# 传入字典进行替换。
df.replace({0:-999, 1:199, 3:399})
```

```python
# 此外还可以区分不同列进行替换，使用正则替换。
```

```python
# replace替换NaN
df.loc[1, 2] = np.nan
display(df, df.replace(np.nan, -999))
```

```python
# map 根据已有列生成新的列。
```

```python
# 适合处理单独的列
df[1].map({0:-999, 1:199, 2:299, 3:399})
```

```python
# map可以传lambda函数、自定义函数
df[3] = df[1].map(lambda x : x+1)
df
```

```python
# 传入自定义方法
def get_num(x):
    res = '未知参数'
    if x == 0:
        res = -777
    elif x == 1:
        res = 177
    elif x == 3:
        res = 377
    return res
df[4] = df[1].map(get_num)
df
```

```python
# 对已有列修改
```

```python
# transform，和map类似，但不仅可以用于Series，而且可以用于DatadFrame
df2 = df.transform(lambda x: x*10)
display(df, df2)
```

```python
# rename专门操作索引的，对索引进行重命名。
df.rename(columns=lambda x : x*10)
```

```python
df.rename(index=lambda x : x*10)
```

### 总结
1. map方法用于将函数或字典应用于Series中的每个元素，进行元素级的映射替换。
2. replace方法用于替换DataFrame或Series中的特定值。它可以替换单个值、列表、字典或使用正则表达式进行替换。
3. transform方法对DataFrame或Series中的数据进行逐元素转换。它可以应用更复杂的函数，包括聚合函数。
4. rename用于替换行索引或列索引。

### 抽样

```python
# 抽样分为两种：有放回抽样，无放回抽样。
df=pd.DataFrame(np.random.randint(0, 150, (30, 3)))
df
```

```python
# 当DataFrame的规模足够大时，直接使用np.random.randint()函数，配合take()函数实现随机抽样。
df.take(np.random.randint(0, len(df), 10))
```

```python
# 无放回抽样：
# permutation函数用于生成不重复的随机数，可以看做对0..n进行随机排列。
df.take(np.random.permutation(len(df))[:10])
```

### sample函数是抽样函数，参数如下
- n: 指定抽取的行数
- frac: 指定抽取的比例(0到1之间)
- replace: 是否有放回抽样，默认False
- weights: 抽样权重
- random_state: 随机种子，保证结果可重现
- axis: 抽样轴向(0为行，1为列)

```python
df.sample(5)
```

```python
df.sample(frac=0.3)
```

```python
df.sample(frac=0.3, replace=True)
```

### 数据聚合
数据分类处理：  
- 分组：先把数据分为几组 group_by
- 聚合：max/min/sum等函数处理
- 合并：一般会把分组聚合后的结果合并到原始DataFrame中。

```python
df = pd.DataFrame({'type':['苹果', '苹果', '栗子', '香蕉', '橘子', '香蕉', '栗子'],
                  'price':np.random.randint(1, 10, 7),
                  'weight':np.random.randint(10, 50, 7)})
df
```

```python
# 按水果类型进行分组
df_g = df.groupby('type')
```

```python
df_g.groups
```

```python
# 对每个类型的水果聚合，计算平均价格、平均重量
df_g_m = df_g.mean()
df_g_m
```

```python
# 计算平均单价。
df_g_m['unit_price'] = df_g_m['price'] / df_g_m['weight']
df_g_m
```

```python
df_unit_price = df_g_m[['unit_price']]
df_unit_price
```

```python
# 聚合结果合并回DataFrame
pd.merge(df, df_unit_price, left_on='type', right_index=True)
```

```python
### 高级数据聚合
```

```python
### 可以使用transfrom和apply进行高级聚合。
df
```

```python
# groupby后直接取其中一列，然后transform进行聚合运算，无需后续的merge合并操作。
df2 = df.copy()
df.groupby(by='type')[['weight']].transform(sum)
df2['weight_sum'] = df.groupby(by='type')[['weight']].transform(sum)
df2
```

```python
# 使用apply函数也有同样的效果。
df3 = df.copy()
df.groupby(by='type')[['weight']].apply(sum)
df3['weight_sum'] = df.groupby(by='type')[['weight']].transform(sum)
df3
```

apply：可以应用于整个行或列，返回标量、Series或DataFrame。

```python
# kde图绘制依赖scipy组件，需要先安装包。
!pip install scipy
```

kde图：核密度估计，用于弥补直方图由于参数bins设置的不合理导致的精度缺失问题
