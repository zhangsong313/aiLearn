KNN（K-Nearest Neighbors，K近邻算法）是机器学习中一种简单而有效的分类和回归算法。

核心思想：如果一个样本在特征空间中的K个最相似（距离最近）的样本中的大多数都属于某一个类别，则该样本也属于这个类别

优点：精度高，异常值不敏感，无数据输入假定.

缺点：时间复杂度和空间复杂度高，对样本不平衡数据集效果差



适用数据范围：标称型，数值型。

```python
# 加载鸢尾花数据集的方法对象。
from sklearn.datasets import load_iris
```

```python
# 划分训练集和测试集的方法对象，后续会用到。
from sklearn.model_selection import train_test_split
```

```python
# 鸢尾花数据集有四个特征列，分别表示花萼长度，花萼宽度，花瓣长度，花瓣宽度。
iris = load_iris()
iris
```

```python
# DESCR可以查看数据集介绍
print(iris.DESCR)
```

```python
# 从数据集对象中获取数据，分类结果，分类名称，特征名称。
data = iris.data
target = iris.target
target_names = iris.target_names
feature_names = iris.feature_names
display(data, target, target_names, feature_names, data.shape)
```

```python
# 整理为DataFrame，查看数据情况
df = pd.DataFrame(data, columns=feature_names)
df.describe()
```

```python
display(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
# 可以看到划分是打乱后划分的，不是直接取最前面或最后面的数据。
y_train
```

```python
# 算法实例和训练
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
```

```python
# 预测
y_ = knn.predict(X_test)
```

```python
# 预测结果和测试结果集对比。
y_ == y_test
```

```python
# True计算时被认为是1，False被认为是0
# 手动评估正确率
(y_ == y_test).mean()
```

```python
# 框架评估正确率
knn.score(X_test, y_test)
```

```python
# 使用训练集评分，不一定百分百正确率。
knn.score(X_train, y_train)
```

找出区分度较大的两个特征。

```python
# 可以看出区分度最明显的是花瓣长度和花瓣宽度。
df.plot(figsize=(8, 6))
```

绘制出2个特征的分布散点图

```python
# 获取最后两列花瓣特征。
data = df.values[:,2:]
data
```

```python
# 使用花瓣两列特征和结果数据画出散点图。
plt.scatter(data[:, 0], data[:, 1], c=target)
```

使用2个特征最大值最小值，获得不超过最值范围内的所有点（横纵1000个）作为测试数据集。

```python
x_max = data[:, 0].max()
x_min = data[:, 0].min()
y_max = data[:, 1].max()
y_min = data[:, 1].min()
display(x_max, x_min, y_max, y_min)
```

```python
# 列出x范围和y范围内的所有点（每行每列1000个点）
x = np.linspace(x_min, x_max, 1000)
y = np.linspace(y_min, y_max, 1000)
display(x, y)
```

```python
# 使用meshgrid方法生成最大值最小值范围内的所有网格点的X和Y坐标
X, Y = np.meshgrid(x, y)
display(X, Y)
```

```python
# ravel可以快速吧ndarray变为一维
# c_可以将x坐标和y坐标组合为二维数组
XY = np.c_[X.ravel(), Y.ravel()]
display(XY, XY.shape)
```

画图，找到knn的分类边界

```python
# 预测出上面二维网格点的结果
knn = KNeighborsClassifier()
knn.fit(data, target)
y_ = knn.predict(XY)
y_
```

```python
# 画图
# 先画网格点的图，否则鸢尾花特征数据点会被遮住。
plt.scatter(XY[:, 0], XY[:, 1], c=y_)
plt.scatter(data[:, 0], data[:, 1], c=target, cmap='rainbow')
```

```python
# 上面一百万个点画起来很耗时。
# pcolormesh可以更快画图
plt.pcolormesh(X, Y, y_.reshape(1000, 1000))
plt.scatter(data[:, 0], data[:, 1], c=target, cmap='rainbow')
```

### knn用于回归

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
# 测试数据
# 随机生成80个数据，范围是0到10
x = np.random.rand(80) * 10
y = np.sin(x)
plt.scatter(x, y)
display(x, y)
```

```python
# 加噪声点
# 每隔4个点，x和y各加上一个标准正态分布。模拟随机噪声点。
# 这里用到了切片后的结果是浅拷贝的原理，累加后会改变原始ndarray
x[::4] += np.random.randn(20)
y[::4] += np.random.randn(20)
plt.scatter(x, y)
```

```python
# 使用回归模型算法训练数据
knn = KNeighborsRegressor()
# 将原始一维数据变形为一列二维数据，进行训练
knn.fit(x.reshape(-1, 1), y)
```

```python
# 使用模型预测数据
X_test = np.linspace(0, 20, 100).reshape(-1, 1)
y_ = knn.predict(X_test)
y_
```

```python
# 绘图
plt.scatter(x, y)
plt.scatter(X_test, y_)
```

可以看出有训练数据的区间，可以进行近似的预测。没有训练数据的区间，预测结果固定是0.

```python
# 评分
y_test = np.sin(X_test)
knn.score(X_test, y_test)
```

回归的评估指标不是准确率，默认的回归评估指标是R^2。
