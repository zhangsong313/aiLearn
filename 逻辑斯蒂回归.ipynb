```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

### sigmoid函数

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
x = np.linspace(-10, 10, 1000)
y = sigmoid(x)
plt.plot(x, y)
```

### 梯度下降函数

```python
def tidu(X, y, learning_rate=0.01, epochs=10000):
    # 样本数和特征数
    m, n = X.shape
    # 初始参数w，b
    w = np.zeros(n)
    b = 0
    # 梯度下降迭代
    loss_his = []
    for i in range(epochs):
        # 线性函数
        z = X @ w + b
        # 概率函数
        y_hat = sigmoid(z)
        # 损失函数
        loss = -1/m * np.sum(y * np.log(y_hat) + (1 - y) * np.log(1 - y_hat))
        # print(f' np.log(y_hat) : { np.log(y_hat)} ')
        loss_his.append(loss)
        # 梯度函数
        dw = 1/m * X.T @ (y_hat - y) 
        db = 1/m * np.sum(y_hat - y)
        # print(f'z : {z} | y_hat : {y_hat} | loss : {loss} | dw : {dw} | db : {db}')
        # 更新参数
        w = w - learning_rate * dw
        b = b - learning_rate * db
    return w, b, loss_his
```

### 预测函数

```python
def predict(X, w, b):
    z = X @ w + b
    y_hat = sigmoid(z)
    return np.where(y_hat >= 0.5, 1, 0)
```

## 作业5：随机梯度下降和小批量梯度下降。

```python
def tidu_SGD(X, y, learning_rate=0.01, epochs=10000):
    # 样本数和特征数
    m, n = X.shape
    # 初始参数w，b
    w = np.zeros(n)
    b = 0
    # 梯度下降迭代
    loss_his = []
    for i in range(epochs):
        # 数据洗牌
        idx = np.random.permutation(len(X))
        X_shuffled = X[idx]
        y_shuffled = y[idx]
        
        # 所有样本损失
        total_loss = 0
        for x_i, y_i in zip(X_shuffled, y_shuffled):
            # 线性函数
            z = x_i @ w + b
            # 概率函数
            y_hat = sigmoid(z)
            # 损失函数
            loss = y_i * np.log(y_hat + 1e-8) + (1 - y_i) * np.log(1 - y_hat + 1e-8)
            total_loss -= loss
            # 梯度函数
            dw = x_i * (y_hat - y_i) 
            db = y_hat - y_i
            # 更新参数
            w = w - learning_rate * dw
            b = b - learning_rate * db
        loss_his.append(total_loss/len(X))
    return w, b, loss_his
```

```python
w, b, loss_his = tidu_SGD(X_train_std, y_train, learning_rate=0.01, epochs=200)
```

```python
display(w, b, loss_his[:5], loss_his[-5:])
```

```python
y_pred = predict(X_test_std, w, b)
(y_pred == y_).mean()
```

```python
# 评分
(y_pred == y_test).mean()
```

```python
# 画出损失下降曲线。
plt.plot(np.arange(len(loss_his)), loss_his)
```

```python
def tidu_MBGD(X, y, learning_rate=0.01, epochs=10000):
    # 样本数和特征数
    m, n = X.shape
    # 初始参数w，b
    w = np.zeros(n)
    b = 0
    # 批量大小
    batch_size = 16
    # 梯度下降迭代
    loss_his = []
    for i in range(epochs):
        # 数据洗牌
        idx = np.random.permutation(len(X))
        X_shuffled = X[idx]
        y_shuffled = y[idx]
        
        # 所有样本损失
        total_loss = 0
        for i in range(0, len(X), batch_size):
            # 当前批次的数据
            X_b = X_shuffled[i:i+batch_size]
            y_b = y_shuffled[i:i+batch_size]
            
            # 线性函数
            z = X_b @ w + b
            # 概率函数
            y_hat = sigmoid(z)
            # 损失函数
            loss = -1 * np.sum(y_b * np.log(y_hat) + (1 - y_b) * np.log(1 - y_hat))
            total_loss += loss
            # 梯度函数
            dw = 1 / batch_size * X_b.T @ (y_hat - y_b) 
            db = 1 / batch_size * np.sum(y_hat - y_b)
            # 更新参数
            w = w - learning_rate * dw
            b = b - learning_rate * db
        loss_his.append(total_loss / len(X))
    return w, b, loss_his
```

```python
w, b, loss_his = tidu_MBGD(X_train_std, y_train, learning_rate=0.01, epochs=200)
```

```python
display(w, b, loss_his[:5], loss_his[-5:])
```

```python
y_pred = predict(X_test_std, w, b)
(y_pred == y_).mean()
```

```python
# 评分
(y_pred == y_test).mean()
```

```python
# 画出损失下降曲线。
plt.plot(np.arange(len(loss_his)), loss_his)
```

### 多输出回归

所有的sklearn模型其实支持多输出回归，也就是每个样本的目标值y不仅是一个值，而是一组值，这样求得的线性方程系数w也不是一维向量，而是二维矩阵。

```python
# 先把数据分割为左半边脸的数据和右半边脸的数据，分别作为特征和目标值。
cols = images.shape[2]
half_cols = int(cols / 2)
left_faces = images[:, :, :half_cols]
right_faces = images[:, :, half_cols:]
```

```python
def showimage(images):
    cols = 10
    rows = int(len(images) / 10)
    fig, axes = plt.subplots(rows, cols, figsize=(cols * 6, rows * 6))
    axes = axes.flatten()
    for i in range(len(images)):
        axes[i].imshow(images[i], cmap='gray')
```
