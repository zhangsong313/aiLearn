**读取adult.txt文件，并使用逻辑斯底回归算法训练模型，根据种族、职业、工作时长来预测一个人的性别**

```python
# 读取数据
df = pd.read_csv("../day28_逻辑回归(三)/代码/data/adults.txt")
df
```

```python
# 二分类问题
df['sex'].unique()
```

```python
# 特征翻译为中文。
df.columns = np.array([  "年龄",  "职业类别",  "最终权重",  "教育程度",  "教育年限",  "婚姻状况",  "职业",  "家庭关系",  "种族",  "性别",  "资本收益",  "资本损失",  "每周工作小时数",  "原籍国家",  "工资"])
df.columns
```

```python
# 过滤本次实验需要的特征和数据。
df_2 = df[["种族", "职业", "每周工作小时数", "性别"]].copy()
df_2
```

```python
# 对字符型字段进行编码，转换为数值型。
encoder = LabelEncoder()
df_2["种族"] = encoder.fit_transform(df_2["种族"])
df_2["职业"] = encoder.fit_transform(df_2["职业"])
df_2
```

```python
# 特征和目标值
data = df_2.iloc[:, :-1].values
target = df_2.iloc[:, -1].values
display(data, target)
```

```python
# 传入数据，进行训练评分。
def fit_score(data, target):
    # 拆分训练集和测试集
    X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=1)
    display(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
    # 训练
    logistic = LogisticRegression()
    logistic.fit(X_train, y_train)
    # 预测
    y_ = logistic.predict(X_test)
    print(y_)
    # 评分
    score = logistic.score(X_test, y_test)
    print(score)
```

```python
fit_score(data, target)
```

**标准化处理**

```python
scaler = StandardScaler()
data_std = scaler.fit_transform(data)
data_std
```

```python
fit_score(data_std, target)
```

**看起来标准化处理对评分没什么影响**

## 作业2 疝气病和马死亡率

**从疝气病症预测病马的死亡率(horseColicTest.txt，horseColicT.txt)**

```python
# 读取数据
df_test = pd.read_csv("../day28_逻辑回归(三)/代码/data/horseColicTest.txt", sep="\t", header=None)
df_train = pd.read_csv("../day28_逻辑回归(三)/代码/data/horseColicTraining.txt", sep="\t", header=None)
display(df_test, df_train)
```

```python
# 二分类问题
display(df_train[21].unique(), df_test[21].unique())
```

```python
X_train = df_train.iloc[:, :-1].values
y_train = df_train.iloc[: , -1].values
X_test = df_test.iloc[:, :-1].values
y_test = df_test.iloc[:, -1].values
display(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
# 传入训练集和测试集数据，训练和评分
def fit_score_horse(X_train, y_train, X_test, y_test, max_iter=100):
    # 训练
    logistic = LogisticRegression(max_iter=max_iter)
    logistic.fit(X_train, y_train)
    # 预测
    y_ = logistic.predict(X_test)
    print(y_)
    # 评分
    score = logistic.score(X_test, y_test)
    print(score)
```

```python
# 迭代次数增加到358才不报错。
fit_score_horse(X_train, y_train, X_test, y_test, max_iter=358)
```

**标准化处理**

```python
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.fit_transform(X_test)
```

```python
fit_score_horse(X_train_std, y_train, X_test_std, y_test, max_iter=17)
```

**标准化处理后迭代次数最小可以调整到17。且评分提高了0.02**

**这说明标准化处理对减小逻辑斯蒂回归迭代次数有帮助。**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

### sigmoid函数

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
x = np.linspace(-10, 10, 1000)
y = sigmoid(x)
plt.plot(x, y)
```

### 梯度下降函数

```python
def tidu(X, y, learning_rate=0.01, epochs=10000):
    # 样本数和特征数
    m, n = X.shape
    # 初始参数w，b
    w = np.zeros(n)
    b = 0
    # 梯度下降迭代
    loss_his = []
    for i in range(epochs):
        # 线性函数
        z = X @ w + b
        # 概率函数
        y_hat = sigmoid(z)
        # 损失函数
        loss = -1/m * np.sum(y * np.log(y_hat) + (1 - y) * np.log(1 - y_hat))
        # print(f' np.log(y_hat) : { np.log(y_hat)} ')
        loss_his.append(loss)
        # 梯度函数
        dw = 1/m * X.T @ (y_hat - y) 
        db = 1/m * np.sum(y_hat - y)
        # print(f'z : {z} | y_hat : {y_hat} | loss : {loss} | dw : {dw} | db : {db}')
        # 更新参数
        w = w - learning_rate * dw
        b = b - learning_rate * db
    return w, b, loss_his
```

### 预测函数

```python
def predict(X, w, b):
    z = X @ w + b
    y_hat = sigmoid(z)
    return np.where(y_hat >= 0.5, 1, 0)
```

### 疝气病和马死亡率案例用梯度下降

```python
# 读取数据
df_test = pd.read_csv("../day28_逻辑回归(三)/代码/data/horseColicTest.txt", sep="\t", header=None)
df_train = pd.read_csv("../day28_逻辑回归(三)/代码/data/horseColicTraining.txt", sep="\t", header=None)
# display(df_test, df_train)
```

```python
X_train = df_train.iloc[:, :-1].values
y_train = df_train.iloc[: , -1].values
X_test = df_test.iloc[:, :-1].values
y_test = df_test.iloc[:, -1].values
# display(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.fit_transform(X_test)
```

```python
# 传入训练集和测试集数据，训练和评分
def fit_score_horse(X_train, y_train, X_test, y_test, max_iter=100):
    # 训练
    logistic = LogisticRegression(max_iter=max_iter)
    logistic.fit(X_train, y_train)
    # 预测
    y_ = logistic.predict(X_test)
    # 评分
    score = logistic.score(X_test, y_test)
    print(f'w: {logistic.coef_} | b : {logistic.intercept_} | y_ : {y_} | score : {score}')
    return y_, score
```

```python
y_, score = fit_score_horse(X_train_std, y_train, X_test_std, y_test, max_iter=17)
```

**试试手动梯度下降的效果**

```python
w, b, loss_his = tidu(X_train_std, y_train, learning_rate=0.1)
```

```python
y_pred = predict(X_test_std, w, b)
(y_pred == y_).mean()
```
