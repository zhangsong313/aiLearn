# 数学（二）

1. 微积分的基本概念
微积分是数学的一个分支，主要研究变化率（导数）和累积量（积分）。
- 导数（Derivative）：描述函数在某一点的变化率
- 积分（Integral）：描述函数在某一区间上的累积效果

2. 导数基础
定义  
导数表示函数 f(x) 在点 x 处的瞬时变化率：  
f'(x) = lim[h→0] [f(x+h) - f(x)] / h


常见导数公式
- 常数函数：(c)' = 0
- 幂函数：(x^n)' = nx^(n-1)
- 三角函数：(sin x)' = cos x，(cos x)' = -sin x
- 指数函数：(e^x)' = e^x


3. 积分基础  
不定积分  
不定积分是导数的逆运算，记作：  
∫ f(x)dx = F(x) + C


其中 F'(x) = f(x)，C 为常数。  
定积分  
定积分表示函数在区间 [a,b] 上的累积量：  
∫[a to b] f(x)dx = F(b) - F(a)


微分：无限小的变化量，变化量趋于无限小时可以忽略，但是因变量的变化量和自变量的变化量比例是可以确定的。  
积分：函数在一段自变量内的累积量。  
求导和积分是互逆运算。
