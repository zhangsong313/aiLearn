```python
# 最大值
# 会偏白
plt.imshow(jin.max(axis=-1), cmap='gray')
```

```python
# 加权平均
# 人眼对于RGB颜色的敏感度不同。
# 根据敏感度加权平均得到灰度值。
plt.imshow(jin.dot([0.299, 0.587, 0.114]), cmap='gray')
```

在 NumPy 中，np.dot() 函数统一处理多种情况：  
一维数组：视为向量点积  
二维数组：视为矩阵乘法  
高维数组：按最后轴进行广义点积运算  

```python
# 数据除以3后显示效果不变。
plt.imshow(jin.dot([0.299, 0.587, 0.114]) / 3, cmap='gray')
```

- matplotlib默认会将数据的最小值映射为黑色，最大值映射为白色
- 中间的值按比例映射到灰度等级
- 因此整体缩放不会改变视觉效果

```python
# 固定显示范围后可以看到数据整体变化对图像显示的影响
plt.imshow(jin.dot([0.299, 0.587, 0.114]), cmap='gray', vmin=0, vmax=1)
```

```python
plt.imshow(jin.dot([0.299, 0.587, 0.114]) / 3, cmap='gray', vmin=0, vmax=1)
```

## 在一个画布中画多张子图

```python
# figure方法先设置画板宽度和高度。
# 单位是英寸(2.5cm左右)
# 每张子图设置大小为4*4，布局是一行三列。
plt.figure(figsize=(4 * 3, 4 * 1))

# 设置画布的布局和子画布的位置 nrows, ncols, index
# 返回当前子画布的坐标轴子图对象
axes1 = plt.subplot(1, 3, 1)
x = np.linspace(-10, 10, 100)
y = np.sin(x)
# 在子图上绘制
axes1.plot(x, y)
axes1.grid(True, ls=':', lw=2)
# x轴和y轴加粗
axes1.axhline(y=0, color='black', lw=2)  # X轴
axes1.axvline(x=0, color='black', lw=2)  # Y轴

# 第二张子图
axes2 = plt.subplot(1, 3, 2)
y2 = np.cos(x)
axes2.plot(x, y2)
axes2.grid(True, ls=':', lw=2)
axes2.axhline(y=0, color='black', lw=2)  # X轴
axes2.axvline(x=0, color='black', lw=2)  # Y轴

# 第三张子图
axes3 = plt.subplot(1, 3, 3)
y3 = x ** 2
axes3.plot(x, y3)
axes3.grid(True, ls=':', lw=2)
axes3.axhline(y=0, color='black', lw=2)  # X轴
axes3.axvline(x=0, color='black', lw=2)  # Y轴
```

```python
# 使用figure.add_subplot也可以绘制子图，而且控制更加精细。
# figure是图形窗口的容器对象，可以理解为画板。包括子图，标题，图例等。
x = np.linspace(-10, 10, 100)
figure = plt.figure(figsize=(4*3, 4*1))
axes1 = figure.add_subplot(1, 3, 1)
axes1.plot(x, np.sin(x))
axes1.grid(True, ls=':', lw=2)
axes1.axhline(y=0, color='black', lw=2)  # X轴
axes1.axvline(x=0, color='black', lw=2)  # Y轴


axes2 = figure.add_subplot(1, 3, 2)
axes2.plot(x, np.cos(x))
axes2.grid(True, ls=':', lw = 2)
axes2.axhline(y=0, color='black',lw=2)
axes2.axvline(x=0, color='black',lw=2)

axes3 = figure.add_subplot(1, 3, 3)
axes3.plot(x, x ** 2)
axes3.grid(True, ls=':', lw=2)
axes3.axhline(y=0, color='black',lw=2)
axes3.axvline(x=0, color='black',lw=2)
```

# matplotlib(三)

单独指定x轴或y轴的界限

```python
# 定义一个方法，将画布设置为数学样式。
def math_style(ax):
    # 隐藏上侧，右侧默认边框
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    # 左侧，下侧坐标轴移动到（0，0）
    ax.spines['left'].set_position(('data', 0))
    ax.spines['bottom'].set_position(('data', 0))
    # 加上坐标轴的箭头
    ax.plot(1, 0, ">", transform=ax.get_yaxis_transform(), clip_on=False, markersize=6)
    ax.plot(0, 1, "^", transform=ax.get_xaxis_transform(), clip_on=False, markersize=6)
    # 设置equal等比例坐标的同时，不修改坐标范围，只拉伸画布。
    ax.set_aspect('equal', adjustable='box') 
    # 网格
    ax.grid(True, alpha=0.3, linestyle='--')
    # x,y轴标签
    ax.set_xlabel('x', loc='right')
    ax.set_ylabel('y', loc='top', rotation=0)
```

```python
x = np.linspace(-10, 10, 100)
y = np.sin(x)
fig, ax = plt.subplots()
ax.plot(x, y)
math_style(ax)

# 使用ylim和xlim指定坐标轴范围
ax.set_ylim((-4, 4))
ax.set_xlim((-3, 3))
```

LaTex语法，显示数学表达式，包括希腊字母等。

```python
# matplotlib支持LaTex语法，以$开始$结束，最好用r，防止转义字符干扰。
x = np.linspace(0, 5, 100)
y = np.sqrt(2/3 * x)

fig, ax = plt.subplots(figsize=(6,4))
ax.plot(x, y, 'r-', linewidth=2)

# 在曲线旁边加公式（LaTeX）
ax.text(1.5, 1.5,  # 坐标位置
        r'$y = \sqrt{\frac{2}{3}x}$',  # 公式
        fontsize=16,
        color='red')

ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$y$')
```

### 三维图

```python
# ---------------------- 1. 定义二元线性方程参数 ----------------------
# z = a*x + b*y + c
a = 0.8
b = 0.6
c = 1.5
```

```python
# ---------------------- 2. 生成网格坐标 ----------------------
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = a * X + b * Y + c  # 二元线性函数
```

```python
# ---------------------- 3. 创建3D画布 ----------------------
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# 绘制平面
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8, edgecolor='none')

# ---------------------- 4. 坐标与标签 ----------------------
ax.set_xlabel('x 轴')
ax.set_ylabel('y 轴')
ax.set_zlabel('z 轴')
ax.set_title(r'$z = ax + by + c$ 二元线性方程 - 3D平面', fontsize=14)

# 保持等比例观感
ax.set_box_aspect([1, 1, 0.8])
```

```python
# 垂直 y、z 轴的平面：x = 固定值
# 这里选 x=3，位置适中、看着最立体
x_fixed = 3

# y 和 z 随便动，形成平面
y = np.linspace(-4, 4, 50)
z = np.linspace(-4, 4, 50)
Y, Z = np.meshgrid(y, z)
X = np.full_like(Y, x_fixed)  # x 固定不变

# 画图
fig = plt.figure(figsize=(9,7))
ax = fig.add_subplot(111, projection='3d')

# 画这张“墙”
ax.plot_surface(X, Y, Z, color='crimson', alpha=0.7)

# 坐标轴
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('垂直于 Y、Z 轴的平面（x=3）', fontsize=14)

# 最佳视角，一眼看出是竖直平面
ax.view_init(elev=20, azim=60)
ax.set_box_aspect([1,1,1])

plt.show()
```
